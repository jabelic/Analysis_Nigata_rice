```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
df_result = pd.read_csv('result.csv', index_col = 0)
print(df_result)
```

```python
month = [i for i in range(1,13)]
PC1 = df_result['PC1'].values
PC2 = df_result['PC2'].values
SVM_score = df_result['SVM正答率'].values
```

```python
fig = plt.figure()
ax = fig.add_subplot(212, projection='3d')
ax.scatter3D(PC1, PC2 ,month )
#plt.legend()
plt.show()

plt.plot(month, PC1, label = "PC1")
plt.legend()
plt.show()
plt.plot(month, PC2 ,label = "PC2", color = "green")
plt.legend()
plt.show()

plt.plot(month, SVM_score ,label = "SVM", color = "red")
plt.legend()
plt.show()
```
